Criação de um climograma para a cidade de Porto Alegre, Brasil
Seria necessária a utilização de outra biblioteca que não estava funcionando. Decidi então, criar o próprio dataFrame utilizando os dados da normal climatológica 1991-2020 do INMET
https://portal.inmet.gov.br/normais
Foram selecionadas as variáveis, Precpitação mensal, temperatura máxima, média e mínima.
Utilizadas as bilbiotecas:
Pandas - criação do DataFrame (de uma planilha de excell
Matplotlib - criação do plot 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arquivo = 'Planilha sem título (1).xlsx'

df = pd.read_excel(arquivo)
df.head()
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.set_ylim([0, 200])
ax2.set_ylim([0, 40])
ax1.set_ylabel('Precptação (mm)', color='black')
ax2.set_ylabel('Temperatura (°C)', color='black')

img1 = ax1.bar(df['Unnamed: 0'] ,df['Precptação mensal'], label='Precptação (mm)')
img2 = ax2.plot(df['Unnamed: 0'] ,df['Temp max'], color='red', label='Temp Max (°C)')
img3 = ax2.plot(df['Unnamed: 0'] ,df['Temp mean'], color='yellow', label='Temp Mean (°C)')
img4 = ax2.plot(df['Unnamed: 0'] ,df['Temp min'], color='black', label='Temp Min (°C)')

plt.suptitle("Climograma Porto Alegre 1991-2020")
plt.legend()
ax1.set_label('Precptação (mm)')
plt.show()



```
